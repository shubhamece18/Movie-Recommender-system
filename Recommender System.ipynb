```python
import pandas as pd
import numpy as np
r_cols=['user_id','movie_id','rating']
ratings=pd.read_csv(r'C:\Users\Aditya\.anaconda\navigator\ml-25m\ratings.csv',names=r_cols,usecols=range(3))

m_cols=['movie_id','title']
movies=pd.read_csv(r'C:\Users\Aditya\.anaconda\navigator\ml-25m\movies.csv',names=m_cols,usecols=range(2))

ratings=pd.merge(movies,ratings)
ratings.head()
```

```python
col=list(ratings.rating)
a=[]
for i in range(len(col)):
    a.append(float(col[i]))
```

```python
ratings.rating=a
ratings.head()
```

```python
movieRatings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating',aggfunc=np.mean)
movieRatings.head()
```

```python
ToyStoryRatings=movieRatings["Toy Story (1995)"]
ToyStoryRatings.head()
```

```python
similarMovies=movieRatings.corrwith(ToyStoryRatings)
similarMovies=similarMovies.dropna()
df=pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
movieStats=ratings.groupby('title').agg({'rating': [np.size,np.mean]})
movieStats.head()
```

```python
popularMovies=movieStats['rating']['size']>=100
movieStats[popularMovies].sort_values([('rating','mean')],ascending=False)[:15]
```

```python
df=movieStats[popularMovies].join(pd.DataFrame(similarMovies,columns=['similarity']))
df.head()
```

```python
df.sort_values(['similarity'],ascending=False)[:15]
```
